### Data Preprocessing

#### Age (Float/Ratio Variable)

```python
#Continuous variables shall be standardized using z-scoring
Age = Selected["Age of Household head"]
Age = pd.DataFrame(Age)

#At first we shall test the skewness of the data through visual inspection
import matplotlib.pyplot as plt
import seaborn as sns
fig, ax = plt.subplots()
sns.histplot(Age, kde=True, ax=ax)
ax.set_title("Age distribution of Household heads",
             fontweight = "bold")
ax.set_ylabel("Frequency")

plt.show()
#The age distribution is fairly symmetric
```

```python
#Now standardizing the age variable through standared scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
Age = scaler.fit_transform(Age)
Age = pd.DataFrame(Age)
Age.columns = ["Age"]
Age.head()
```

#### Age below 5 years (Float/Ratio Variable)

```python
#The continuous variable (of count nature) will be standardized appropriately
import numpy as np
Below_5 = Selected["Number of members below 5 years"]
Below_5 = pd.DataFrame(Below_5)
Below_5.replace(np.nan, 0, inplace = True)

import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots()
sns.histplot(Below_5, kde = True, ax=ax)
ax.set_ylabel("Frequency")
plt.show()
#This count data shows significant skewness towards right towards zero
```

```python
#Log plus one standardization technique shall be used given the high density of zero count in the dataset. Addition of one is essential for
#each value because it can help treat the issue of performing log on zero value, which is essentially undefined and interfers with 
# log-transform standardization

import numpy as np
Below_5 = np.log1p(Below_5)
Below_5.head()
```

```python
#Now transforming the OLR result summary into dataframe for further manipulation, filtering based on given criteria, and exporting to CSV 
results_html = res.summary().tables[1].as_html()
results_degree = pd.read_html(results_html, header = 0, index_col = 0)[0] #excerpted the scraping work from Gemini
```

```python
#Now generating the list demonstrating the relationship between exo and endogenous variables 
Relationship = []
for values in results_degree["coef"]:
    if values <= 0:
        results = "negative"
    else:
        results = "positive"
    Relationship.append(results) 

#Turning the list into a dataframe
Relationship = pd.DataFrame(Relationship)
Relationship.columns = ["Relationship"]

#Now merging the dataframe with the OLR results
results_degree.reset_index(inplace = True)
results_degree = pd.merge(results_degree, Relationship, how = "left", left_index= True, right_index = True)
results_degree.set_index("index", inplace = True)
results_degree.to_csv("results_degree.csv")
```
